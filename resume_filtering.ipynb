```python


import os
import re
import threading
import fitz  # PyMuPDF
import tkinter as tk
from tkinter import filedialog, messagebox
from tkinter.scrolledtext import ScrolledText
from tkinter import ttk

# ---------------------------- Core Logic ----------------------------

def extract_text_from_pdf(file_path: str) -> str:
    """Extract text from a PDF using PyMuPDF. Returns empty string if unreadable."""
    try:
        doc = fitz.open(file_path)
        parts = []
        for page in doc:
            parts.append(page.get_text("text"))
        doc.close()
        return "\n".join(parts)
    except Exception:
        return ""

def clean_keywords(raw: str):
    """Split comma-separated keywords and clean whitespace."""
    return [k.strip() for k in raw.split(",") if k.strip()]

def find_matched_keywords(text: str, keywords):
    """Whole-word/phrase case-insensitive matching for each keyword."""
    matched = []
    for kw in keywords:
        # Whole-word for single words; phrase match for multi-word terms
        if " " in kw:
            pattern = re.escape(kw)
        else:
            pattern = r"\b" + re.escape(kw) + r"\b"
        if re.search(pattern, text, flags=re.IGNORECASE):
            matched.append(kw)
    return matched

def scan_resumes(folder: str, keywords, mode: str):
    """
    Returns a list of results:
    (filename, score%, matched_keywords_list, missing_keywords_list)
    """
    pdf_files = [f for f in os.listdir(folder) if f.lower().endswith(".pdf")]
    results = []

    for f in pdf_files:
        path = os.path.join(folder, f)
        text = extract_text_from_pdf(path)

        if not text.strip():
            # PDF unreadable
            results.append((f, 0, [], ["(PDF read error)"]))
            continue

        matched = find_matched_keywords(text, keywords)
        missing = [k for k in keywords if k not in matched]

        if mode == "ALL" and missing:
            # Not eligible in ALL mode
            score = 0
        else:
            score = int((len(matched) / max(1, len(keywords))) * 100)

        results.append((f, score, matched, missing))

    # Sort by score desc, then name
    results.sort(key=lambda x: (-x[1], x[0].lower()))
    return results, len(pdf_files)

# ---------------------------- UI Helpers ----------------------------

def set_status(msg: str):
    status_var.set(msg)
    root.update_idletasks()

def browse_folder():
    folder_path = filedialog.askdirectory(title="Select Resume Folder")
    if folder_path:
        folder_var.set(folder_path)

def open_selected_pdf(event=None):
    # Open the selected resume in default PDF viewer
    selection = results_list.curselection()
    if not selection:
        return
    idx = selection[0]
    name = results_list.get(idx).split("  |  ")[0].strip()
    folder = folder_var.get().strip()
    path = os.path.join(folder, name)

    if os.path.exists(path):
        try:
            os.startfile(path)  # Windows
        except Exception:
            messagebox.showinfo("Open File", f"Could not auto-open file.\nPath:\n{path}")
    else:
        messagebox.showwarning("Missing File", "File not found on disk.")

def run_filter():
    folder = folder_var.get().strip()
    keywords = clean_keywords(keyword_var.get())
    mode = mode_var.get()

    if not folder or not os.path.isdir(folder):
        messagebox.showwarning("Input Error", "Please select a valid resume folder.")
        return
    if not keywords:
        messagebox.showwarning("Input Error", "Please enter at least one keyword.")
        return

    # UI disable during scan
    run_btn.config(state="disabled")
    browse_btn.config(state="disabled")
    set_status("Scanning resumes... please wait.")
    progress.start(10)

    # clear outputs
    results_list.delete(0, tk.END)
    details_text.config(state="normal")
    details_text.delete("1.0", tk.END)
    details_text.config(state="disabled")

    def worker():
        try:
            results, total = scan_resumes(folder, keywords, mode)
        except Exception as e:
            results, total = [], 0
            err = str(e)
        else:
            err = None

        def update_ui():
            progress.stop()
            run_btn.config(state="normal")
            browse_btn.config(state="normal")

            if err:
                set_status("Error occurred.")
                messagebox.showerror("Error", err)
                return

            set_status(f"Done. Scanned {total} PDF resumes.")

            # populate list
            for name, score, matched, missing in results:
                # show top summary line
                results_list.insert(tk.END, f"{name}  |  Score: {score}%  |  Matched: {len(matched)}")

            # store latest results in app state
            app_state["results"] = results
            app_state["keywords"] = keywords
            app_state["mode"] = mode

        root.after(0, update_ui)

    threading.Thread(target=worker, daemon=True).start()

def on_select_result(event=None):
    selection = results_list.curselection()
    if not selection:
        return

    idx = selection[0]
    results = app_state.get("results", [])
    keywords = app_state.get("keywords", [])

    if idx >= len(results):
        return

    name, score, matched, missing = results[idx]

    details = []
    details.append(f"Resume: {name}")
    details.append(f"Score: {score}%")
    details.append(f"Mode: {app_state.get('mode', 'ANY')}")
    details.append("")
    details.append(f"Keywords Entered ({len(keywords)}):")
    details.append(", ".join(keywords))
    details.append("")
    details.append(f"Matched Keywords ({len(matched)}):")
    details.append(", ".join(matched) if matched else "None")
    details.append("")
    details.append(f"Missing Keywords ({len(missing)}):")
    details.append(", ".join(missing) if missing else "None")
    details.append("")
    details.append("Tip: Double-click the result to open the PDF.")

    details_text.config(state="normal")
    details_text.delete("1.0", tk.END)
    details_text.insert(tk.END, "\n".join(details))
    details_text.config(state="disabled")

# ---------------------------- Professional UI ----------------------------

root = tk.Tk()
root.title("Resume Filtering System (Professional UI)")
root.geometry("980x620")
root.minsize(900, 580)

# Theme + styling (ttk)
style = ttk.Style()
style.theme_use("clam")

BG = "#0f172a"
CARD = "#111c33"
CARD2 = "#0b1220"
TXT = "#e5e7eb"
MUTED = "#94a3b8"
ACCENT = "#38bdf8"
GREEN = "#22c55e"

root.configure(bg=BG)

style.configure("TFrame", background=BG)
style.configure("Card.TFrame", background=CARD)
style.configure("Card2.TFrame", background=CARD2)

style.configure("TLabel", background=BG, foreground=TXT, font=("Segoe UI", 10))
style.configure("Title.TLabel", background=BG, foreground=TXT, font=("Segoe UI", 20, "bold"))
style.configure("Sub.TLabel", background=BG, foreground=MUTED, font=("Segoe UI", 10))

style.configure("Card.TLabel", background=CARD, foreground=TXT, font=("Segoe UI", 10))
style.configure("CardTitle.TLabel", background=CARD, foreground=TXT, font=("Segoe UI", 12, "bold"))
style.configure("Card2.TLabel", background=CARD2, foreground=TXT, font=("Segoe UI", 10))

style.configure("TEntry", padding=6)
style.map("TButton", relief=[("pressed", "sunken"), ("!pressed", "raised")])

# top header
header = ttk.Frame(root)
header.pack(fill="x", padx=18, pady=(16, 8))

ttk.Label(header, text="üìÑ Resume Filtering System", style="Title.TLabel").pack(anchor="w")
ttk.Label(
    header,
    text="Filter PDF resumes using keywords ‚Ä¢ Match ANY or Match ALL ‚Ä¢ Shows match score and details",
    style="Sub.TLabel"
).pack(anchor="w", pady=(4, 0))

# Main body container
body = ttk.Frame(root)
body.pack(fill="both", expand=True, padx=18, pady=12)

# Left: Controls card
left = ttk.Frame(body, style="Card.TFrame", padding=18)
left.pack(side="left", fill="y", padx=(0, 12))

ttk.Label(left, text="Controls", style="CardTitle.TLabel").pack(anchor="w", pady=(0, 12))

# Folder
folder_var = tk.StringVar()
ttk.Label(left, text="Resume Folder", style="Card.TLabel").pack(anchor="w")
folder_entry = ttk.Entry(left, textvariable=folder_var, width=42)
folder_entry.pack(fill="x", pady=(6, 8))

browse_btn = ttk.Button(left, text="üìÇ Browse Folder", command=browse_folder)
browse_btn.pack(fill="x", pady=(0, 14))

# Keywords
keyword_var = tk.StringVar()
ttk.Label(left, text="Keywords (comma-separated)", style="Card.TLabel").pack(anchor="w")
kw_entry = ttk.Entry(left, textvariable=keyword_var, width=42)
kw_entry.pack(fill="x", pady=(6, 14))

# Mode
mode_var = tk.StringVar(value="ANY")
ttk.Label(left, text="Matching Mode", style="Card.TLabel").pack(anchor="w")
mode_row = ttk.Frame(left, style="Card.TFrame")
mode_row.pack(anchor="w", pady=(8, 14))

ttk.Radiobutton(mode_row, text="Match ANY", variable=mode_var, value="ANY").pack(side="left", padx=(0, 12))
ttk.Radiobutton(mode_row, text="Match ALL", variable=mode_var, value="ALL").pack(side="left")

# Run button
run_btn = ttk.Button(left, text="üîç Run Filter", command=run_filter)
run_btn.pack(fill="x", pady=(4, 10))

# Progress + status
progress = ttk.Progressbar(left, mode="indeterminate")
progress.pack(fill="x", pady=(6, 10))

status_var = tk.StringVar(value="Ready.")
status_label = ttk.Label(left, textvariable=status_var, style="Card.TLabel")
status_label.pack(anchor="w", pady=(2, 0))

# Tips
ttk.Label(left, text="Tips:", style="CardTitle.TLabel").pack(anchor="w", pady=(18, 8))
tips = (
    "‚Ä¢ Use keywords like: Python, SQL, Machine Learning\n"
    "‚Ä¢ Use commas between keywords\n"
    "‚Ä¢ Double-click a result to open the PDF\n"
    "‚Ä¢ ALL mode returns only resumes matching all keywords"
)
ttk.Label(left, text=tips, style="Card.TLabel").pack(anchor="w")

# Right side: Results + Details
right = ttk.Frame(body)
right.pack(side="left", fill="both", expand=True)

# Results card
results_card = ttk.Frame(right, style="Card.TFrame", padding=16)
results_card.pack(fill="both", expand=True, pady=(0, 12))

ttk.Label(results_card, text="Results", style="CardTitle.TLabel").pack(anchor="w", pady=(0, 10))

# Listbox with scrollbar
list_frame = ttk.Frame(results_card, style="Card.TFrame")
list_frame.pack(fill="both", expand=True)

scroll_y = tk.Scrollbar(list_frame)
scroll_y.pack(side="right", fill="y")

results_list = tk.Listbox(
    list_frame,
    yscrollcommand=scroll_y.set,
    bg="#071023",
    fg=TXT,
    selectbackground=ACCENT,
    selectforeground="#02111d",
    font=("Segoe UI", 10),
    relief="flat",
    highlightthickness=1,
    highlightbackground="#1f2a44",
    activestyle="none"
)
results_list.pack(side="left", fill="both", expand=True)
scroll_y.config(command=results_list.yview)

results_list.bind("<<ListboxSelect>>", on_select_result)
results_list.bind("<Double-Button-1>", open_selected_pdf)

# Details card
details_card = ttk.Frame(right, style="Card2.TFrame", padding=16)
details_card.pack(fill="x")

ttk.Label(details_card, text="Details", style="CardTitle.TLabel").pack(anchor="w", pady=(0, 8))

details_text = ScrolledText(
    details_card,
    height=8,
    bg="#020617",
    fg=TXT,
    insertbackground=TXT,
    font=("Consolas", 10),
    relief="flat",
    highlightthickness=1,
    highlightbackground="#1f2a44"
)
details_text.pack(fill="both", expand=True)
details_text.config(state="disabled")

# App state
app_state = {
    "results": [],
    "keywords": [],
    "mode": "ANY",
}

set_status("Ready. Select a folder and enter keywords.")
root.mainloop()
```
